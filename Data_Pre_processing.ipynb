# Imputation Techniques

1. **Mean Value Imputation** :
here we will try to replace the missing value, by the finding the mean of that sepcific column

```python
sns.displot(df['age'])
```

```python
sns.distplot(df['age'])
```

```python
sns.histplot(df['age'],kde=True) # we are having some 177 missing values
```

```python
df['Age_mean']=df['age'].fillna(df['age'].mean())
```

```python
df[['Age_mean','age']]
```

so, we can see in the above part the NaN values are getting replaced with mean age values.

# IMP : Mean value imputation works well when we have data in the normal distribution or gaussian distribution aspect.

# 2. Median Value Imputation ::
useful when the data has outliers in the dataset or data is skewed.

```python
df['Age_median']=df['age'].fillna(df['age'].median())
```

```python
df[['Age_mean','Age_median','age']][0:25]
```

# Handling Imbalanced Dataset

By these techniques:
1. Up Sampling

2. Down Sampling

```python
import numpy as np

import pandas as pd

np.random.seed(123)

n_samples=1000

class_0_ratio=0.9
n_class_0=int(n_samples*class_0_ratio)
n_class_1=n_samples-n_class_0
```

```python
n_class_0,n_class_1
```

```python
# creating a dataframe with imbalanced dataset

class_0 = pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})

class_1=pd.DataFrame({
    'feature_1': np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature_2': np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

# 1. Upsampling


```python
df_minority=df[df['target']==1]

df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample

# this library helps to create more points for the minority points

df_minority_upsampled=resample(df_minority,replace=True,
         n_samples=len(df_majority),
         random_state=42
         )
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
```

```python
df_upsampled['target'].value_counts()
```
