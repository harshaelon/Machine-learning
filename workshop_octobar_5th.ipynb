# outlier detection

```python
# z-score
l1 =  [1,2,3,4,7,5,11,13,15,3,2,1]
# IQR
```

```python
import numpy as np

d1 = [1,2,1,2,1,1,2,2,15,17,6,5,7,2,2,1]

d2 = [1,1,2,2,1,40,1,1,2,2]

# calculating the mean

mean1 = np.mean(d2)
sd = np.std(d2)

print(mean1,sd)
```

```python
# z-score formula : element - mean / s.d


outlier = []
threshold = 3
for i in d2:
  z_score=(i-mean1)/sd
  if(z_score>threshold):
    outlier.append(i)
```

```python
outlier
```

```python
# IQR

l1 = [-15,-4,-7,-1,1,2,3,2,2,1,1,15,2,3,23,17,9,44]

l2 = np.sort(l1)
```

```python
q1 = np.percentile(l1 , 25 , interpolation = 'midpoint')

q2 = np.percentile(l1, 50 , interpolation = 'midpoint')

q3 = np.percentile(l1 , 75 , interpolation = 'midpoint')
```

```python
IQR = q3-q1

print(IQR)
```

```python
up_limit = IQR*1.5 + q3

low_limit = IQR*1.5 - q1

ll2 = q1 - 1.5*IQR
ul2 = q3 + 1.5*IQR

print(ll2,ul2)
print(up_limit,low_limit)
```

```python
for i in l1:
  if( (i > ul2) or (i<ll2)):
    print(i)
```
